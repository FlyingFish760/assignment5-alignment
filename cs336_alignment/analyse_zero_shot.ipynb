```python
import json
# Evaluate the overall model performance based the result

data_path = r"../data/sft-cs336-assign5-datasets/sft-reason/sft_gpt-oss-120b_filtered.jsonl"
res_path = r"./zero_shot_performance.jsonl"

total_format_score, total_answer_score, total_reward_score = 0, 0, 0
with open(res_path, "r", encoding="utf-8") as f:
    for line in f:
        res = json.loads(line)
        total_format_score += res["eval_score"]["format_reward"]
        total_answer_score += res["eval_score"]["answer_reward"]
        total_reward_score += res["eval_score"]["reward"]

with open(data_path, "r", encoding='utf-8') as f:
    data = json.load(f)
num_samples = len(data)
print(f"Total format score is {total_format_score} / {num_samples}.")
print(f"Total answer score is {total_answer_score} / {num_samples}.")
print(f"Total reward score is {total_reward_score} / {num_samples}.")
```

```python
# Sample results on different conditions
format_wrong = []
answer_wrong = []
correct = []
with open(res_path, "r", encoding="utf-8") as f:
    for line in f:
        res = json.loads(line)
        eval_score = res["eval_score"]
        if eval_score["format_reward"] == 0:
            format_wrong.append(res)
        elif eval_score["answer_reward"] == 0:
            answer_wrong.append(res)
        elif eval_score["reward"] == 1:
            correct.append(res)
```

```python
for i, sample in enumerate(format_wrong):
    if i >= 10: break
    print(sample['generated_text'])
    print('----------------------------')
```

```python
for i, sample in enumerate(answer_wrong):
    if i >= 10: break
    print("prompt:", sample["prompt"])
    print("response:", sample["generated_text"])
    print('----------------------------')
```

```python
for i, sample in enumerate(correct):
    if i >= 10: break
    print("prompt:", sample["prompt"])
    print("response:", sample["generated_text"])
    print('----------------------------')
```

结果：
对于
1）format wrong：一定是model的response不符合prompt模板，属于base model的问题
2）answer wrong：有些是model answer错误，有些则是model answer正确，但reward_func解析错误问题
3）correct：通过几个correct的样本，可以发现answer判定为correct的样本，其answer格式一定为<answer>数字(标点)</answer>
